```python
# --- HARMONIZAÇÃO E SOBRESCRITA DAS RESPOSTAS DE 2017 ---

# 1. Escolaridade
mapa_esc_2017 = {'A': 'A', 'B': 'A', 'C': 'A', 'D': 'B', 'E': 'C', 'F': 'D', 'G': 'E'}
# Sobrescreve as colunas originais com os novos valores
df2017_3['Q_ESCOLARIDADE_MAE'] = df2017_3['Q_ESCOLARIDADE_MAE'].map(mapa_esc_2017)
df2017_3['Q_ESCOLARIDADE_PAI'] = df2017_3['Q_ESCOLARIDADE_PAI'].map(mapa_esc_2017)

# 2. Mora com os pais
mapa_mora_com = {'A': 'B', 'B': 'A', 'C': 'B'}
# Sobrescreve as colunas originais
df2017_3['Q_MORA_COM_MAE'] = df2017_3['Q_MORA_COM_MAE'].map(mapa_mora_com)
df2017_3['Q_MORA_COM_PAI'] = df2017_3['Q_MORA_COM_PAI'].map(mapa_mora_com)

# 3. Bens domésticos
mapa_itens = {'A':'A', 'B':'B', 'C':'C', 'D':'D', 'E':'D'}
colunas_bens = [
    'Q_BENS_TV_CORES', 'Q_BENS_GELADEIRA', 'Q_BENS_CARRO',
    'Q_BENS_COMPUTADOR', 'Q_BENS_BANHEIRO', 'Q_BENS_QUARTOS'
    # Adicione as outras colunas de bens se desejar
]
for coluna in colunas_bens:
    # Sobrescreve a coluna original dentro do loop
    df2017_3[coluna] = df2017_3[coluna].map(mapa_itens)

# 4. Tempo de lazer e trabalho doméstico
mapa_tempo = {'E':'A', 'A':'B', 'B':'C', 'C':'D', 'D':'D'}
# Sobrescreve as colunas originais
df2017_3['Q_LAZER'] = df2017_3['Q_LAZER'].map(mapa_tempo)
df2017_3['Q_TRABALHO_DOMESTICO'] = df2017_3['Q_TRABALHO_DOMESTICO'].map(mapa_tempo)

print("Todas as colunas foram transformadas e sobrescritas com sucesso!")
```

```python
import pandas as pd
import numpy as np

# --- PASSO 1: PADRONIZAR VALORES FALTANTES ---
valores_para_substituir = ['.', '*', np.nan]
print("Padronizando valores faltantes para '*' em todos os DataFrames...")
df2017_3 = df2017_3.replace(valores_para_substituir, '*')

nome_arquivo_final = 'SAEB_UNIFICADO_2017.csv'
print(f"\nExportando para '{nome_arquivo_final}'...")
df2017_3.to_csv(nome_arquivo_final, index=False, sep=';', encoding='utf-8-sig')
print("🎉 Processo concluído com sucesso!")
```
